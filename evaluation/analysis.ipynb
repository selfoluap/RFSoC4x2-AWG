```python
import pandas as pd


data = pd.read_csv('serrodyne_data.csv')
data_precorrected = pd.read_csv('serrodyne_data_precorrection.csv')
```

```python
from matplotlib import pyplot as plt

plt.plot(data['Frequency'], data['Magnitude'], label='Without Predistortion')
plt.plot(data_precorrected['Frequency'], data_precorrected['Magnitude'], label='With Predistortion')
plt.xlabel('Frequency (MHz)')
plt.ylabel('Magnitude (dBm)')
plt.title('Serrodyne Frequency vs Magnitude')
plt.legend()
plt.grid()
plt.show()
```
